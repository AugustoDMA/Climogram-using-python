Este roteiro tem como objetivo guiar os estudantes para a construção de cinco climogramas para cinco cidades de diferentes regiões do país com o uso da linguagem de programação "Python".



Primeiramente, é interessante entender o que compõe um climograma:

 Nada mais é que um gráfico com o propósito de representar as variações de temperatura e precipitação de uma localidade ao longo do ano

 São de grande ajuda para a identificação do clima das regiões, exibindo seus padrões.

 Sua forma básica é de qualquer outro gráfico, com eixos x e y, como o gerado com o código abaixo:

```python
#BIBLIOTECAS PARA OS DADOS (São nescessarias somente no primeiro código)
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D


# Configuração do gráfico - tamanho e quantidade de graficos
fig, ax1 = plt.subplots(figsize=(10, 6))

# Ajustes visuais - Nome do gráfico e fonte
plt.title('Esqueleto climograma', fontweight='bold', pad=20)

# Mostrar o gráfico
plt.show()
```

Como queremos 5 climogramas, alteramos o código para que ele gere mais gráficos:

```python


# Configuração da figura com 5 subplots (gráficos)
fig, axs = plt.subplots(1, 5, figsize=(30, 6))  # 1 linha, 5 colunas
fig.suptitle('Esqueleto climogramas', fontweight='bold', y=1.02)  # Título principal

# Ajustar layout para evitar sobreposição
plt.tight_layout()
plt.show()
```

Podemos organizar em colunas ao em vez de somente uma linha:

```python
fig, axs = plt.subplots(3, 2, figsize=(20, 15))
plt.subplots_adjust(wspace=0.3, hspace=0.4)
fig.suptitle('Climogramas de 5 Cidades Brasileiras',
             fontsize=16, fontweight='bold', y=0.98)

fig.delaxes(axs[2, 1]) # Para remover o gráfico excedente

plt.tight_layout()
plt.show()
```

Agora, adicionamos as linhas de código que nos permitem adicionar os dados nestes gráficos. Mas onde adquirimos estes dados?

No site do INMET (Instituto Nacional de Meteorologia - https://portal.inmet.gov.br/), seguir os seguintes passos:

No menu “Clima”, selecionar “Normais climatológicas” e em seguida “Downloads” Selecionar o período escolhido e as seguintes planilhas para download:

-Temperatura média compensada mensal e anual (°C)

-Precipitação acumulada mensal e anual das 5 cidades escolhidas.

Segue o exemplo de um climograma de uma cidade qualquer:

```python


# Vamo dalhe
LOCALIDADE = "Cidade ficticia"  # Nome do local estudado (insira entre as aspas)
MESES = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
         'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

# Insira seus dados aqui (12 valores para cada variável):
TEMPERATURA = [30, 23.0, 22.0, 20.0, 17.5, 16.0,
               15.5, 16.5, 17.5, 19.0, 20.5, 21.5]  # em °C

PRECIPITACAO = [230, 200, 170, 80, 70, 50,
                40, 50, 80, 120, 150, 200]  # em mm


#Estética dos gráficos

fig, ax = plt.subplots(figsize=(12, 6)) #Quantidade e tamanho de gráficos

# Barras de precipitação (eixo direito)
ax.bar(MESES, PRECIPITACAO, color='#1E88E5', alpha=0.7, width=0.6, label='Precipitação')
ax.set_ylabel('Precipitação (mm)', color='#1E88E5', fontsize=12)
ax.tick_params(axis='y', labelcolor='#1E88E5')

# Linha de temperatura (eixo esquerdo)
ax_temp = ax.twinx()
ax_temp.plot(MESES, TEMPERATURA, color='#D81B60', marker='o',
             linewidth=2, markersize=8, label='Temperatura')
ax_temp.set_ylabel('Temperatura (°C)', color='#D81B60', fontsize=12)
ax_temp.tick_params(axis='y', labelcolor='#D81B60')

# Ajustes estéticos (Titulo e localização das legendas)
plt.title(f'Climograma - {LOCALIDADE}\n',
          fontsize=14, fontweight='bold', pad=20)
ax.set_xlabel('Meses do Ano', fontsize=12)
ax.set_xticks(range(len(MESES)))
ax.grid(axis='x', alpha=0.3)

# Unificar legendas
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax_temp.get_legend_handles_labels()
ax_temp.legend(lines + lines2, labels + labels2,
               loc='upper right', framealpha=1)

plt.tight_layout()
plt.show()

```

Agora, o código adaptado para a geração dos 5 climogramas:

```python
#Troque pelos seus dados obtidos no INMET! (Cuidado com as vírgulas e colchetes.)
dados = {
    "São Paulo": {
        "temp": [22.5, 23.0, 22.0, 20.0, 17.5, 16.0, 15.5, 16.5, 17.5, 19.0, 20.5, 21.5],
        "precip": [240, 200, 170, 80, 70, 50, 40, 50, 80, 120, 150, 200]
    },
    "Rio de Janeiro": {
        "temp": [26.5, 26.8, 26.0, 24.5, 22.8, 21.5, 21.0, 21.5, 22.0, 23.0, 24.5, 25.5],
        "precip": [180, 150, 140, 100, 80, 60, 50, 60, 90, 110, 130, 170]
    },
    "Belo Horizonte": {
        "temp": [23.0, 23.5, 23.0, 21.5, 19.0, 17.5, 17.0, 18.5, 20.0, 21.5, 22.0, 22.5],
        "precip": [280, 220, 180, 60, 30, 20, 15, 25, 50, 120, 190, 250]
    },
    "Augusto Pestana": {
        "temp": [26.0, 26.0, 26.0, 26.5, 26.5, 26.5, 26.5, 27.0, 27.0, 27.0, 26.5, 26.5],
        "precip": [260, 280, 310, 300, 260, 120, 80, 70, 100, 130, 180, 220]
    },
    "Porto Alegre": {
        "temp": [24.5, 24.0, 22.5, 19.5, 16.0, 14.0, 13.5, 15.0, 16.5, 19.0, 21.0, 23.0],
        "precip": [120, 110, 100, 90, 100, 110, 120, 130, 140, 130, 110, 100]
    }
}

MESES = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']


fig, axs = plt.subplots(3, 2, figsize=(20, 15))
plt.subplots_adjust(wspace=0.3, hspace=0.4)
fig.suptitle('Climogramas de 5 Cidades Brasileiras',
             fontsize=16, fontweight='bold', y=0.98)

# Cores e limites padronizados
COR_PRECIP = '#1E88E5'
COR_TEMP = '#D81B60'
LIMITE_PRECIP = 350 #Estes valores padronizam os eixos dos climogramas
LIMITE_TEMP_MIN = 10
LIMITE_TEMP_MAX = 30

# Elementos para a legenda
legend_elements = [
    Line2D([0], [0], color=COR_PRECIP, lw=4, label='Precipitação (mm)'),
    Line2D([0], [0], color=COR_TEMP, marker='o', lw=2, label='Temperatura (°C)', markersize=8)
]

# Plotar cada cidade
for idx, (cidade, valores) in enumerate(dados.items()):
    row = idx // 2
    col = idx % 2
    ax = axs[row, col]

    # Barras de precipitação
    bars = ax.bar(MESES, valores["precip"], color=COR_PRECIP, alpha=0.7, width=0.6)
    ax.set_title(cidade, fontweight='bold', pad=20)
    ax.set_ylabel('Precipitação (mm)', color=COR_PRECIP)
    ax.tick_params(axis='y', labelcolor=COR_PRECIP)
    ax.set_xticks(range(len(MESES)))
    ax.set_xticklabels(MESES, rotation=45)
    ax.set_ylim(0, LIMITE_PRECIP)

    # Linha de temperatura
    ax_temp = ax.twinx()
    line = ax_temp.plot(MESES, valores["temp"], color=COR_TEMP,
                        marker='o', linewidth=2, markersize=5)
    ax_temp.set_ylabel('Temperatura (°C)', color=COR_TEMP)
    ax_temp.tick_params(axis='y', labelcolor=COR_TEMP)
    ax_temp.set_ylim(LIMITE_TEMP_MIN, LIMITE_TEMP_MAX)

    # Adicionar legenda dentro de cada gráfico
    ax.legend(handles=legend_elements, loc='upper right', framealpha=1)

# Remover o subplot vazio e ajustar layout
fig.delaxes(axs[2, 1])


plt.tight_layout()
plt.show()
```
